## ADD SUBSTRACT MUTLIPLY AND DIVIDE

```python
s1= p.Series(data= [1,5,3])
s2= p.Series(data= [4,5,6])
print(f'Addition= \n{s1+s2}')
print(f'substraction= \n{s2-s1}')
print(f'Multiplication= \n{s1*s2}')
print(f'Division= \n{s1/s1}')
```

## COMPARE TWO PANDA SERIES

```python
s1.compare(s2,keep_shape=True)
```

```python
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],'score': [12.5, 9, 16.5, n.nan, 9, 20, 14.5, n.nan, 8, 19],'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1], 'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
df= p.DataFrame(exam_data)
df
```

## SELECT SPECIFIED COLUMNS AND ROWS

```python
df.iloc[0:3,0:2]
```

## SELECT SCORES BETWEEN 15 AND 20

```python
df[(df.score>=15) & (df.score<=20)]
```

## NUMBER OF ATTEMPTS LESS THAN 2 AND SCORE MORE THAN 15

```python
df[(df.attempts<2) & (df.score>15)]
```

## APPEND A NEW ROW AND THEN DELETE IT

```python
new= {'name':'Arpan','score':10,'attempts': 1, 'qualify':'no'}
df = df.append(new,ignore_index= True)
df
```

```python
df= df.drop([10])
df
```
